```python
# 튜플,딕셔너리 공부내용
# Immutable(변하지 않는) : 정수, 실수, 문자열, 튜플
# Mutable(변하기 쉬운) : 리스트, 딕셔너리, 집합

a=[1,2,3] #리스트: 값 변경가능
b=(1,2,3) #튜플: 값 변경불가능 (고정됨)
c={
    "이름": "김민수",
    "나이": "23",
    "학교": "KPU"
} #딕셔너리
print(c["이름"])

c["집"]="안양" #이렇게 추가 가능
print(c)
del c["집"] #이렇게 삭제 가능
print(c)

d={1:"a",1:"b"}
print(d) #키가 중복될 경우 뒤에것만 남음
d={1:"a",2:"a"}
print(d) #값이 같은건 쌉가능

a={1:"a", 2:"b", 3:"c"}
print(a.keys()) #키값만 빼기
print(a.values()) #값만 빼기
print(a.items()) #둘다 빼기

for x in a.keys():
    print(x) #키값만 빼는 반복문
for x,y in a.items():
    print(x) #키도 빼고
    print(y) #값도 빼고

a.clear() #a 비우기
print(a)

a={1:"a", 2:"b", 3:"c"}

print(a[1]) # 키에따른 값이 없으면 오류
print(a.get(4)) # 없는값은 None으로 출력
print(a.get(4,"없음")) #이렇게 하면 키에따른 값이 없을때 뒤에 써놓은값 출력

print(4 in a) #False 출력
print(1 in a) #True 출력
```

```python
# 클래스
# 함수를 여러개 찍어내기 위해서 사용
class Calculator:         # 클래스 이름의 시작은 대문자
    def __init__(self,first,second):   # 초기값 설정
        self.first=first
        self.second=second
    def add(self):
        return self.first+self.second
    
cal1=Calculator(3,5)
cal2=Calculator(4,6)
print(cal1.add())
print(cal2.add())

# 클래스의 상속
class Calculatorplus(Calculator): # Claculator가 상속됨
    def sub(self):
        return self.first-self.second
a=Calculatorplus(4,3)             # 기존의 클래스를 활용하는 방법
print(a.add())                    # __init__ 초기값도 그대로 적용
print(a.sub())

# 클래스 오버라이딩
class Safecal(Calculator):
    def sub(self):                 # 기존 함수기능 추가가능
        if self.second < 0:        # 상속받은 자식 우선시
            return '굳이???'
        else:
            return self.first-self.second
a=Safecal(4,-3)
print(a.sub())
a=Safecal(4,3)
print(a.sub())
```
